```python
import os
import re
from pathlib import Path

import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from tqdm.notebook import tqdm

import albumentations as A
import segmentation_models_pytorch as smp

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
```

```python
CFG = {
    "data_root": r"C:\Users\ASHAH\Desktop\DL_Project\breast_ultrasound\combined_breast_ultrasound_dataset",
    "img_size": 320,
    "batch_size": 8,
    "epochs": 25,
    "lr": 1e-3,
    "num_workers": 0,
    "encoder_name": "timm-efficientnet-b3",   # fallback: "efficientnet-b3"
    "threshold": 0.5,
}
```

```python
DATASET_ROOT = Path(CFG["data_root"])

def base_key(filename: str) -> str:
    name = filename.replace(".png", "")
    name = re.sub(r"_mask(_\d+)?$", "", name)
    return name

rows = []
for cls_name in ["benign", "malignant", "normal"]:
    cls_dir  = DATASET_ROOT / cls_name
    img_dir  = cls_dir / "images"
    mask_dir = cls_dir / "masks"

    # Strip _mask suffix from mask stem to get the base key
    # e.g. "uclm_benign_0452_mask" → "uclm_benign_0452"
    masks = {
        re.sub(r"_mask$", "", p.stem): str(p)
        for p in mask_dir.glob("*.png")
    }

    for img_path in sorted(img_dir.glob("*.png")):
        key = img_path.stem  # e.g. "busi_benign_0001"
        rows.append({
            "image_path": str(img_path),
            "mask_path":  masks.get(key, None),  # now matches correctly
            "label":      cls_name,
        })

df = pd.DataFrame(rows)
print(df["label"].value_counts())
print("Total:", len(df))

# Verify matching worked — count how many images have a mask
print("With mask:", df["mask_path"].notna().sum())
print("Without mask:", df["mask_path"].isna().sum())
```

```python
train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    random_state=42,
    stratify=df["label"]
)

valid_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    random_state=42,
    stratify=temp_df["label"]
)

train_df = train_df.reset_index(drop=True)
valid_df = valid_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

print("Train:", len(train_df))
print(train_df["label"].value_counts(), "\n")

print("Valid:", len(valid_df))
print(valid_df["label"].value_counts(), "\n")

print("Test:", len(test_df))
print(test_df["label"].value_counts())
```

```python
train_ds_noaug = BUSISegDataset(train_df, transforms=train_tfms_noaug)
valid_ds_noaug = BUSISegDataset(valid_df, transforms=valid_tfms_noaug)
test_ds_noaug  = BUSISegDataset(test_df,  transforms=valid_tfms_noaug)

train_loader_noaug = DataLoader(
    train_ds_noaug,
    batch_size=CFG["batch_size"],
    shuffle=True,
    num_workers=CFG["num_workers"],
    pin_memory=True
)

valid_loader_noaug = DataLoader(
    valid_ds_noaug,
    batch_size=CFG["batch_size"],
    shuffle=False,
    num_workers=CFG["num_workers"],
    pin_memory=True
)

test_loader_noaug = DataLoader(
    test_ds_noaug,
    batch_size=CFG["batch_size"],
    shuffle=False,
    num_workers=0,
    pin_memory=True
)
```

```python
# Fresh model for no-aug experiment
model_noaug = smp.DeepLabV3Plus(
    encoder_name=CFG["encoder_name"],
    encoder_weights='imagenet',
    in_channels=1,
    classes=1,
    activation=None
).to(device)

# Freeze encoder initially
for param in model_noaug.encoder.parameters():
    param.requires_grad = False
print("Encoder frozen ✓")

optimizer_noaug = torch.optim.AdamW(model_noaug.parameters(), lr=CFG["lr"], weight_decay=1e-2)
scaler_noaug    = torch.amp.GradScaler("cuda", enabled=torch.cuda.is_available())
scheduler_noaug = torch.optim.lr_scheduler.CosineAnnealingLR(
    optimizer_noaug, T_max=CFG["epochs"], eta_min=1e-6
)
```

```python
best_dice_noaug = -1
no_improve_noaug = 0
patience = 10

history_noaug = {
    "train_loss": [],
    "train_dice": [],
    "val_loss":   [],
    "val_dice":   []
}

for epoch in range(CFG["epochs"]):
    if epoch == 5:
        for param in model_noaug.encoder.parameters():
            param.requires_grad = True
        print("Encoder unfrozen ✓")

    train_loss, train_dice = train_one_epoch(model_noaug, train_loader_noaug)
    val_loss,   val_dice   = valid_one_epoch(model_noaug, valid_loader_noaug)
    scheduler_noaug.step()

    history_noaug["train_loss"].append(train_loss)
    history_noaug["train_dice"].append(train_dice)
    history_noaug["val_loss"].append(val_loss)
    history_noaug["val_dice"].append(val_dice)

    print(f"\nEpoch {epoch+1}/{CFG['epochs']}")
    print(f"Train Loss: {train_loss:.4f} | Train Dice: {train_dice:.4f}")
    print(f"Val   Loss: {val_loss:.4f} | Val   Dice: {val_dice:.4f}")
    print(f"LR: {scheduler_noaug.get_last_lr()[0]:.2e}")

    if val_dice > best_dice_noaug:
        best_dice_noaug = val_dice
        no_improve_noaug = 0
        torch.save(model_noaug.state_dict(), "best_noaug_model.pth")
        print("Saved best_noaug_model.pth ✓")
    else:
        no_improve_noaug += 1
        if no_improve_noaug >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break
```

```python
train_tfms = A.Compose([
    A.Resize(CFG["img_size"], CFG["img_size"]),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.3),
    A.ShiftScaleRotate(
        shift_limit=0.05,
        scale_limit=0.10,
        rotate_limit=15,
        border_mode=cv2.BORDER_CONSTANT,
        p=0.5
    ),
    A.ElasticTransform(p=0.3),        # ← simulates tissue deformation
    A.GridDistortion(p=0.3),          # ← common in medical imaging
    A.RandomBrightnessContrast(p=0.4), # ← handles ultrasound variability
    A.GaussianBlur(p=0.2),            # ← simulates noise
    A.GaussNoise(p=0.2),              # ← ultrasound speckle noise
])
```

```python
class BUSISegDataset(Dataset):
    def __init__(self, df, transforms=None):
        self.df = df.reset_index(drop=True)
        self.transforms = transforms

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        image = cv2.imread(row["image_path"], cv2.IMREAD_GRAYSCALE)
        if image is None:
            raise FileNotFoundError(row["image_path"])

        mask_path = row["mask_path"]
        if mask_path and Path(mask_path).exists():
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            mask = (mask > 0).astype(np.float32)
        else:
            mask = np.zeros(image.shape, dtype=np.float32)

        # ← Resize BOTH to same size before albumentations
        h, w = CFG["img_size"], CFG["img_size"]
        image = cv2.resize(image, (w, h), interpolation=cv2.INTER_LINEAR)
        mask  = cv2.resize(mask,  (w, h), interpolation=cv2.INTER_NEAREST)

        if self.transforms is not None:
            transformed = self.transforms(image=image, mask=mask)
            image = transformed["image"]
            mask  = transformed["mask"]

        image = image.astype(np.float32) / 255.0
        image = np.expand_dims(image, axis=0)
        mask  = np.expand_dims(mask.astype(np.float32), axis=0)

        return {
            "image": torch.tensor(image, dtype=torch.float32),
            "mask":  torch.tensor(mask,  dtype=torch.float32),
            "label": row["label"],
        }
```

```python
#DataLoader
train_ds = BUSISegDataset(train_df, transforms=train_tfms)
valid_ds = BUSISegDataset(valid_df, transforms=valid_tfms)

train_loader = DataLoader(
    train_ds,
    batch_size=CFG["batch_size"],
    shuffle=True,
    num_workers=CFG["num_workers"],
    pin_memory=True
)

valid_loader = DataLoader(
    valid_ds,
    batch_size=CFG["batch_size"],
    shuffle=False,
    num_workers=CFG["num_workers"],
    pin_memory=True
)
```

```python
#Initialize model
model = smp.DeepLabV3Plus(
    encoder_name="timm-efficientnet-b3",
    encoder_weights='imagenet',     # FROM SCRATCH
    in_channels=1,
    classes=1,
    activation=None
).to(device)

print(type(model))
```

```python
#Loss func
dice_loss = smp.losses.DiceLoss(mode="binary", from_logits=True)
bce_loss = nn.BCEWithLogitsLoss()

def compute_loss(logits, masks):
    return dice_loss(logits, masks) + bce_loss(logits, masks)
```

```python
#Dice metrics
def dice_score_from_logits(logits, masks, threshold=0.5, eps=1e-7):
    preds = (torch.sigmoid(logits) > threshold).float()
    inter = (preds * masks).sum(dim=(1,2,3))
    union = preds.sum(dim=(1,2,3)) + masks.sum(dim=(1,2,3))
    dice = (2 * inter + eps) / (union + eps)
    return dice.mean().item()
```

```python
#Optimizer and amp
optimizer = torch.optim.AdamW(model.parameters(), lr=CFG["lr"])
scaler = torch.amp.GradScaler("cuda", enabled=torch.cuda.is_available())
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
    optimizer,
    T_max=CFG["epochs"],   # decays over full training
    eta_min=1e-6           # minimum LR it decays to
)
```

```python
#Train loop
def train_one_epoch(model, loader):
    model.train()
    running_loss = 0.0
    running_dice = 0.0

    pbar = tqdm(loader, total=len(loader))
    for batch in pbar:
        images = batch["image"].to(device, non_blocking=True)
        masks = batch["mask"].to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)

        with torch.amp.autocast("cuda", enabled=torch.cuda.is_available()):
            logits = model(images)
            loss = compute_loss(logits, masks)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item()
        running_dice += dice_score_from_logits(logits.detach(), masks)

        pbar.set_postfix({
            "loss": f"{running_loss/(pbar.n+1):.4f}",
            "dice": f"{running_dice/(pbar.n+1):.4f}"
        })

    return running_loss / len(loader), running_dice / len(loader)
```

```python
#Valid loop
@torch.no_grad()
def valid_one_epoch(model, loader):
    model.eval()
    running_loss = 0.0
    running_dice = 0.0

    for batch in tqdm(loader, total=len(loader)):
        images = batch["image"].to(device, non_blocking=True)
        masks = batch["mask"].to(device, non_blocking=True)

        with torch.amp.autocast("cuda", enabled=torch.cuda.is_available()):
            logits = model(images)
            loss = compute_loss(logits, masks)

        running_loss += loss.item()
        running_dice += dice_score_from_logits(logits, masks)

    return running_loss / len(loader), running_dice / len(loader)
```

```python
best_dice = -1
patience = 10        # ← stop if no improvement for 10 epochs
no_improve = 0

history = {
    "train_loss": [],
    "train_dice": [],
    "val_loss":   [],
    "val_dice":   []
}

for epoch in range(CFG["epochs"]):
    if epoch == 5:
        for param in model.encoder.parameters():
            param.requires_grad = True
        print("Encoder unfrozen ✓")

    train_loss, train_dice = train_one_epoch(model, train_loader)
    val_loss, val_dice     = valid_one_epoch(model, valid_loader)
    scheduler.step()

    history["train_loss"].append(train_loss)
    history["train_dice"].append(train_dice)
    history["val_loss"].append(val_loss)
    history["val_dice"].append(val_dice)

    print(f"\nEpoch {epoch+1}/{CFG['epochs']}")
    print(f"Train Loss: {train_loss:.4f} | Train Dice: {train_dice:.4f}")
    print(f"Val   Loss: {val_loss:.4f} | Val   Dice: {val_dice:.4f}")
    print(f"LR: {scheduler.get_last_lr()[0]:.2e}")

    if val_dice > best_dice:
        best_dice = val_dice
        no_improve = 0
        torch.save(model.state_dict(), "best_deeplabv3plus_busi_scratch.pth")
        print("Saved best model ✓")
    else:
        no_improve += 1
        if no_improve >= patience:
            print(f"Early stopping at epoch {epoch+1} — no improvement for {patience} epochs")
            break
```
